# Objective:

Draw —by hand or using computer— the vector force field of the tidal force given by Eq. 5.51. An approximation is presented in Eqs. 5.54

## Background

From Eq. 5.51, the tidal force $F_{T}$ on mass m on Earth's surface is given by:

\begin{align}
\mathbf{F}_{T} = -GmM_{M} \bigg(\frac{\mathbf{e}_{R}}{R^{2}}-\frac{\mathbf{e}_{D}}{D^{2}}\bigg) \tag{5.51}
\end{align}

where $G$ is the Gravitational Constant; $m$ is the point mass of any object placed along the surface of the Earth; $M_{m}$ is the Moon's mass; $D$ is the distance between the Earth and the Moon; $R$ is the distance between the point mass and the Moon, $\mathbf{e}_{R}$ is the unit vector pointing towards $m$, along $R$; and $\mathbf{e}_{D}$ is the unit vector pointing towards Earth, along $D$

```python
from IPython.display import Image
Image(filename="Tidal Vectors from Eq. 5.51 plot.png") 
```

An approximation using binomial expansion on Eq. 5.51 gives the x- and y-components of the tidal force:

\begin{align}
F_{T_{x}} = \frac{2GmM_{M}r\cos\theta}{D^{3}}  \tag{5.54a}
\end{align}

\begin{align}
F_{T_{y}} =-\frac{GmM_{M}r\sin\theta}{D^{3}}  \tag{5.54b}
\end{align}

which we shall use to numerically plot the 2D vector field provided by Eq. 5.51.

## Plotting the Vector Field

```python
from scipy.constants import G
G
```

to prevent matplotlib error, set unit of length to km

```python
G = G/(0.001**3)
G
```

We use matplotlib's quiver on a polar plot to produce the vector force field. We set the value of the small mass $m$ to 1 to simplify the coding process.

```python
M_m = 7.348*10**19 #Moon Mass in kg
D = 384400 #Earth-Moon Distance in km
r = 6371 #Earth radius in km

radii = np.linspace(0,r,8)
thetas = np.linspace(0,2*np.pi)
theta, R = np.meshgrid(thetas, radii)

u = (2*G*M_m*R*np.cos(theta))/D**3 #5.54a
v = (-G*M_m*R*np.sin(theta))/D**3 #5.54b

f = plt.figure(figsize=(7, 7))
ax = f.add_subplot(111, polar=True)

ax.quiver(theta, R, u, v)
plt.polar(np.linspace(-np.pi,np.pi,100),np.linspace(r,r,100), c='blue' ) #Earth circle

ax.add_artist(plt.Circle((0, 0), r, transform=ax.transData._b, fill=True, linewidth=2, alpha=0.2)) #filled circle

ax.set_title("Vector Force Field of the Tidal Force Along Earth's Surface" + "\n" +"(Polar View)", size=20)

ax.axis('off')
ax.set_aspect(1.0)
ax.set_rlim(0, 8000)
ax.set_xticklabels([])
ax.grid(False)
```
