## Background

From Eq. 5.51, the tidal force $F_{T}$ on mass m on Earth's surface is given by:

\begin{align}
\mathbf{F}_{T} = -GmM_{M} \bigg(\frac{\mathbf{e}_{R}}{R^{2}}-\frac{\mathbf{e}_{D}}{D^{2}}\bigg) \tag{5.51}
\end{align}

where $G$ is the Gravitational Constant; $m$ is the point mass of any object placed along the surface of the Earth; $M_{m}$ is the Moon's mass; $D$ is the distance between the Earth and the Moon; $R$ is the distance between the point mass and the Moon, $\mathbf{e}_{R}$ is the unit vector pointing towards $m$, along $R$; and $\mathbf{e}_{D}$ is the unit vector pointing towards Earth, along $D$

```python
from matplotlib.patches import Ellipse, Arc, FancyArrowPatch

Ocean_rad = 2
Moon_rad = 0.5
Moon_dist_x = -8
Moon_dist_y = 0

x_Ocean = Ocean_rad*np.cos(np.linspace(-np.pi,np.pi,100))
y_Ocean = Ocean_rad*np.sin(np.linspace(-np.pi,np.pi,100))

x_Moon = Moon_rad*np.cos(np.linspace(-np.pi,np.pi,100))+Moon_dist_x
y_Moon = Moon_rad*np.sin(np.linspace(-np.pi,np.pi,100))+Moon_dist_y


fig, ax = plt.subplots(figsize=(15,10))
ax.set_title("Tidal Vectors from Eq. 5.51", {'fontsize':15})
ax.set_aspect("equal")

#Earth Ocean
ax.plot(x_Ocean,y_Ocean, c="blue")
ax.add_artist(plt.Circle((0,0),Ocean_rad,color="blue", alpha=0.2))

#Moon
ax.plot(x_Moon,y_Moon, c="grey")
ax.add_artist(plt.Circle((Moon_dist_x, Moon_dist_y),0.5,color="grey", alpha=0.2))

Tf1_arrow = FancyArrowPatch((-2,0),
                           (-3 , 0), 
                           arrowstyle="Simple,tail_width=0.5,head_width=4,head_length=8")
ax.add_patch(Tf1_arrow)
ax.annotate('$\mathbf{F}_{T}$', xy=(0,0), 
            xytext=(-3,-0.5))

Tf2_arrow = FancyArrowPatch((2,0),
                           (3 , 0), 
                           arrowstyle="Simple,tail_width=0.5,head_width=4,head_length=8")
ax.add_patch(Tf2_arrow)
ax.annotate('$\mathbf{F}_{T}$', xy=(0,0), 
            xytext=(2.5,-0.5))

eD_arrow = FancyArrowPatch((-7.5,0),
                           (-6.5 , 0), 
                           arrowstyle="Simple,tail_width=0.5,head_width=4,head_length=8")
ax.add_patch(eD_arrow)
ax.annotate('$\mathbf{e}_{D}$', xy=(0,0), 
            xytext=(-7.5,-0.5))

ax.plot((-2,2),(0,0),c="black", alpha=0.5)
ax.plot((0,0),(-2,2),c="black", alpha=0.5)

ax.plot((-8,0),(0,0),c="black", alpha=0.5)
ax.plot((-8,0),(0,2),c="black", alpha=0.5)

ax.plot(0, 2, marker='o', markersize=3, color="black")
ax.annotate('$m$', xy=(0,0), 
            xytext=(-0.5,2.1))
ax.annotate('$D$', xy=(-4,0), 
            xytext=(-4,-0.5))
ax.annotate('$R$', xy=(-4,0), 
            xytext=(-4,1.2))

eR1_arrow = FancyArrowPatch((0,2),
                           (+1.5*8/np.sqrt(8**2+2**2) , 2+1.5*2/np.sqrt(8**2+2**2)), 
                           arrowstyle="Simple,tail_width=0.5,head_width=4,head_length=8")
ax.add_patch(eR1_arrow)
ax.annotate('$\mathbf{e}_{R}$', xy=(0,0), 
            xytext=(1,2.5))

eR2_arrow = FancyArrowPatch((0,-2),
                           (+1.5*8/np.sqrt(8**2+2**2) , -2-1.5*2/np.sqrt(8**2+2**2)), 
                           arrowstyle="Simple,tail_width=0.5,head_width=4,head_length=8")
ax.add_patch(eR2_arrow)
ax.annotate('$\mathbf{e}_{R}$', xy=(0,0), 
            xytext=(1,-2.7))

ax.axis('off')
ax.set_xlim(-10,10)
ax.set_ylim(-4,4)
```

An approximation using binomial expansion on Eq. 5.51 gives the x- and y-components of the tidal force:

\begin{align}
F_{T_{x}} = \frac{2GmM_{M}r\cos\theta}{D^{3}}  \tag{5.54a}
\end{align}

\begin{align}
F_{T_{y}} =-\frac{GmM_{M}r\sin\theta}{D^{3}}  \tag{5.54b}
\end{align}

which we shall use to numerically plot the 2D vector field provided by Eq. 5.51.

## Plotting the Vector Field

```python
from scipy.constants import G
G
```

to prevent matplotlib error, set unit of length to km

```python
G = G/(0.001**3)
G
```

We use matplotlib's quiver on a polar plot to produce the vector force field. We set the value of the small mass $m$ to 1 to simplify the coding process.

```python
M_m = 7.348*10**19 #Moon Mass in kg
D = 384400 #Earth-Moon Distance in km
r = 6371 #Earth radius in km

radii = np.linspace(0,r,8)
thetas = np.linspace(0,2*np.pi)
theta, R = np.meshgrid(thetas, radii)

u = (2*G*M_m*R*np.cos(theta))/D**3 #5.54a
v = (-G*M_m*R*np.sin(theta))/D**3 #5.54b

f = plt.figure(figsize=(7, 7))
ax = f.add_subplot(111, polar=True)

ax.quiver(theta, R, u, v)
plt.polar(np.linspace(-np.pi,np.pi,100),np.linspace(r,r,100), c='blue' ) #Earth circle

ax.add_artist(plt.Circle((0, 0), r, transform=ax.transData._b, fill=True, linewidth=2, alpha=0.2)) #filled circle

ax.set_title("Vector Force Field of the Tidal Force Along Earth's Surface" + "\n" +"(Polar View)", size=20)

ax.axis('off')
ax.set_aspect(1.0)
ax.set_rlim(0, 8000)
ax.set_xticklabels([])
ax.grid(False)
```
